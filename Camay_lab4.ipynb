```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('movie_metadata_cleaned.csv')

x = df["director_facebook_likes"]
y = df["gross"]

df
```

```python

df[['director_facebook_likes', 'gross']]
```

```python

def correlation(x,y):

    corr_ans = x.corr(y)
    print('Correlation : ',corr_ans)
    if corr_ans > 0:
        print('Correlation is positive')
    else:
        print('Correlation is negative')
   
    sns.regplot(x, y)

 

correlation(x,y)
```

```python
a = df['budget']
b = df.plot_keywords.str.count('\|') +1 
b = b.replace(1,0).values


def covariance(a,b):

    covar_ans = np.cov(a,b, bias=True)[0,1]
    stdx = np.std(a)
    stdy = np.std(b)
    corr3 = covar_ans /(stdx*stdy)
    print('Covariance : ', corr3)
    
def plot(a,b):
    sns.regplot(b,a )
   
covariance(a,b)
plot(a,b)
```
